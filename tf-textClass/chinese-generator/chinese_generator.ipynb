```python
'''
    文本预处理
'''

input_path = "../dataset/chinese.txt"
output_path = "../dataset/chinese_process.txt"

pre_text = open(input_path,'r').read()
after_text = open(output_path,'w')

for line in pre_text:
    after_text.write(line.strip())
```

```python
# 文本分词
text = open(input_path,'r').read()
train_data = []
vocab_text = []
for line in text:
    print(line)
    train_data.append(["BOS"] + [c for c in jieba.__lcut(line)] + ["EOS"])
    for c in jieba.__lcut(line):
         vocab_text.append(c)

# 词典构建
vocab = sorted(set(vocab_text))
print(len(vocab))
print(vocab)

```
